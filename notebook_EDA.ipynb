## 1- INTRODUCCIÓN: VARIABLES A UTILIZAR

#### **1. Life Ladder**
Descripción: El "Life Ladder" (Escalera de la Vida) es una medida de bienestar subjetivo.
Escala: 0 a 10.

Interpretación:
0: Representa la peor vida posible según la percepción del individuo.
10: Representa la mejor vida posible según la percepción del individuo.

Cálculo: Se basa en la pregunta de la encuesta Gallup World Poll donde se pide a los encuestados que imaginen una escalera con peldaños numerados del 0 al 10 y que indiquen en qué peldaño se sienten actualmente.

#### **2. GDP per Capita**
Descripción: Producto Interno Bruto (PIB) per cápita ajustado por paridad de poder adquisitivo (PPA).

Medición: Se mide en dólares internacionales ajustados por PPA.

Cálculo: Utiliza datos del Banco Mundial y otras fuentes económicas para calcular el PIB per cápita, ajustando las diferencias en el costo de vida entre países.

#### **3. Social Support**
Descripción: Medida de la percepción de apoyo social.

Interpretación: Escala del 0 al 1

Cálculo: Basada en la pregunta de la encuesta Gallup World Poll: "Si te encuentras en dificultades, ¿tienes familiares o amigos con quienes puedas contar para recibir ayuda cuando la necesites, o no?". La proporción de personas que responden 
"sí" se utiliza para esta variable.

#### **4. Healthy Life Expectancy**
Descripción: Esperanza de vida saludable al nacer.

Medición: Años.

Cálculo: Combina datos de la Organización Mundial de la Salud (OMS) y otras fuentes de salud pública para calcular los años de vida esperados en buena salud.


#### **5. Freedom to Make Life Choices**
Descripción: Medida de la percepción de libertad para tomar decisiones sobre la vida.

Interpretación: Escala del 0 al 1

Cálculo: Basada en la pregunta de la encuesta Gallup World Poll: "¿Estás satisfecho o insatisfecho con tu libertad para 
elegir lo que haces con tu vida?". La proporción de personas que responden "satisfecho" se utiliza para esta variable.

#### **6. Generosity**
Descripción: Medida de la generosidad de la población.

Interpretación: Escala del 1 al -1
Cálculo: Basada en la pregunta de la encuesta Gallup World Poll: "¿Has donado dinero a una organización benéfica en el último mes?". La proporción de personas que responden "sí" se ajusta por el PIB per cápita.


#### **7. Perceptions of Corruption**
Descripción: Medida de la percepción de corrupción en el gobierno y negocios.

Interpretación: Escala del 0 al 1 (como más cerca de 1 más corrupción percibida)

Cálculo: Basada en la pregunta de la encuesta Gallup World Poll: "¿Está extendida la corrupción en el gobierno o no?" y "¿Está extendida la corrupción en los negocios o no?". La proporción de personas que responden "sí" a cualquiera de las preguntas se utiliza para esta variable.


#### **8. Positive Affect**
Descripción: Medida de las emociones positivas.

Interpretación: Escala del 0 al 1

Cálculo: Basada en preguntas de la encuesta Gallup World Poll sobre si los encuestados experimentaron ciertas emociones positivas (risa, disfrute, etc.) el día anterior a la encuesta. La proporción de respuestas positivas se utiliza para esta variable.


#### **9. Negative Affect**
Descripción: Medida de las emociones negativas.

Interpretación: Escala del 0 al 1

Cálculo: Basada en preguntas de la encuesta Gallup World Poll sobre si los encuestados experimentaron ciertas emociones negativas (preocupación, tristeza, etc.) el día anterior a la encuesta. La proporción de respuestas positivas se utiliza para esta variable.

##  **5-AÑADIR VARIABLES NECESARIAS**

```python
# Para añadir variable CONTINENTE

# Función para obtener el nombre del continente a partir del nombre del país
def continentes(country_name):
    try:
        country_code = pycountry.countries.lookup(country_name).alpha_2
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except LookupError:
        return None

# Agregar una nueva columna 'continent' con el nombre del continente
df_felicidad['continente'] = df_felicidad['Country name'].apply(continentes)
```

El **Banco Mundial** clasifica a los países en función de su PIB per cápita ajustado por PPA. En 2023, la clasificación del Banco Mundial para los ingresos es la siguiente:

· **Ingresos Altos**: PIB per cápita superior a $13,205 USD.

Ingresos Medios:

----**Medios Altos**: PIB per cápita entre $4,256 y $13,205 USD.

----**Medios Bajos**: PIB per cápita entre $1,046 y $4,255 USD.

· **Ingresos Bajos**: PIB per cápita inferior a $1,046 USD.

```python
# Para dejar de tener un logaritmo y tener el PIB per cápita real de cada país y cada año
# Nos servirá para clasificar los países después y crear una nueva variable
df_felicidad['PIB per capita'] = np.exp(df_felicidad['Log GDP per capita'])
```

```python
def categoria_ingresos(pib_per_capita):
    if pib_per_capita > 13205:
        return 'Ingresos Altos'
    elif 4257 <= pib_per_capita <= 13205:
        return 'Ingresos Medios Altos'
    elif 1046 <= pib_per_capita <= 4256:
        return 'Ingresos Medios Bajos'
    else:
        return 'Ingresos Bajos'

df_felicidad['categoria de ingresos'] = df_felicidad['PIB per capita'].apply(categoria_ingresos)
```

```python
df_felicidad = df_felicidad.drop(columns=['Income Category'])
```

```python
def clasificar_periodo(year):
    if year <= 2019:
        return 'PRE-COVID'
    elif year >= 2020:
        return 'POST COVID'
    else:
        return 'No Definido'

# Añadir la nueva columna 'Periodo' al DataFrame
df_felicidad['Periodo'] = df_felicidad['year'].apply(clasificar_periodo)
```

##  **6-ANÁLISIS UNIVARIANTE**


```python
# Usamos la siguiente función para sacar de cada variable los valores estadísticos, la distribución y el boxplot.
# Eso nos ayuda a saber, sobre todo, como se distribuye nuestra muestra y cuales son sus valores estadísticos principales
#En la función hay una figura para cada variable con los dos subgráficos
def analisis_univariante(df, column):
    print(f"Análisis univariante para {column}:") #con el describe vemos los principales estadísticos
    print(df[column].describe())
    
    # Crear una figura con dos subgráficos (distribución muestra y boxplot)
    plt.figure(figsize=(14, 6))
    
    # Primer subgráfico: Histograma con KDE(Kernel Density Estimation)
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribución de {column}')
    
    # Segundo subgráfico: Boxplot sobre todo para ver valores extremos
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot de {column}')
    
    # Ajustar el layout y mostrar la figura
    plt.tight_layout()
    plt.show()

# Columnas para el análisis univariante
columns = df_felicidad.columns

# Usamos la función para ver el análisis univariante planteado en todas las variables
for col in columns:
    if df_felicidad[col].dtype in ['float64', 'int64']:
        analisis_univariante(df_felicidad, col)
```

### Buscamos saber ahora los países más felices, es decir, ver que es lo que ha ido pasando en los últimos años en este world happiness report


```python
def mas_menos_felices(df, start_year, end_year):
    # Iterar sobre cada año en el rango especificado
    for year in range(start_year, end_year + 1):
        # Filtrar los datos para el año actual
        filtered_df = df[df['year'] == year]
        
        if filtered_df.empty:
            print(f"No hay datos para el año {year}.")
            continue
        
        # Obtener los valores máximos y mínimos de 'Life Ladder' por país
        max_life_ladder_per_country = filtered_df.groupby('Country name')['Life Ladder'].max().reset_index()
        min_life_ladder_per_country = filtered_df.groupby('Country name')['Life Ladder'].min().reset_index()
        
        # Renombrar las columnas para claridad
        max_life_ladder_per_country.rename(columns={'Life Ladder': 'Max Life Ladder'}, inplace=True)
        min_life_ladder_per_country.rename(columns={'Life Ladder': 'Min Life Ladder'}, inplace=True)
        
        # Merge para tener una única tabla con maximos y mínimos
        result = pd.merge(max_life_ladder_per_country, min_life_ladder_per_country, on='Country name', how='outer')
        
        # Añadir columna de año
        result['Year'] = year
        
        # Ordenar el resultado por 'Max Life Ladder' en orden descendente
        result_sorted = result.sort_values(by='Max Life Ladder', ascending=False)
        
        # Mostrar el resultado para el año actual
        print(f"\nResultados para el año {year}:")
        print(result_sorted[['Country name', 'Year', 'Max Life Ladder', 'Min Life Ladder']])
```

## **7- ANÁLISIS BIVARIANTE**

```python
# Planteamos una función que permita sacar un scatter plot de todas las variables con la variable de life ladder o indice de felicidad
# Lo ponemos todo en la misma figura. Se puede observar la variable life ladder en el eje y siempre
# Nos sirve para entender bien como se relacionan todas las variables con la que ponemos de objetivo: Life ladder

def analisis_bivariante(df, target, columns):
    num_vars = len(columns)
    
    # Creamos figura con subgráficos
    plt.figure(figsize=(18, 6 * num_vars))
    
    for i, col in enumerate(columns):
        if col != target:
            plt.subplot(num_vars, 1, i + 1)
            sns.scatterplot(data=df, x=col, y=target)
            plt.title(f'Relación entre {col} y {target}')
            plt.xlabel(col)
            plt.ylabel(target)
    
    # Añadimos y ajustamos el layout
    plt.tight_layout()
    plt.show()

# Definimos la variable objetivo: Life ladder
objetivo = 'Life Ladder'

# Introducimos un filtro para que coja todas las variables numéricas exceptuando la objetivo y lo hacemos todo una lista
columns = df_felicidad.columns[df_felicidad.dtypes != 'object'].tolist()
columns.remove(objetivo)

# Pintamos las gráficas
analisis_bivariante(df_felicidad, objetivo, columns)
```

```python
def matriz_correlacion_continente(df, continente=None):

    if continente:
        # Filtra el DataFrame por el continente especificado
        df_continente = df[df['continente'] == continente]
        
        # Verifica si hay datos para el continentee especificado
        if df_continente.empty:
            raise ValueError(f"No hay datos para el continente '{continente}'")
        
        # Selecciona las columnas numéricas que queremos usar (excluyendo 'Country name' y 'year')
        columns = ['Life Ladder', 'Log GDP per capita', 'Social support', 
                   'Healthy life expectancy at birth', 'Freedom to make life choices', 
                   'Generosity', 'Perceptions of corruption', 
                   'Positive affect', 'Negative affect']
        
        # Asegúrate de que todas las columnas numéricas existan en el DataFrame filtrado
        if all(col in df_continente.columns for col in columns):
            # Calcula las correlaciones de 'Life Ladder' con todas las demás variables
            correlations = df_continente[columns].corr().loc['Life Ladder']
            
            # Convierte a formato largo para facilitar la visualización
            correlations = correlations.reset_index()
            correlations.columns = ['Variable', 'Correlation']
            correlations['continente'] = continente
            
            return correlations
        else:
            raise ValueError("Algunas columnas necesarias no están presentes en el DataFrame.")
    else:
        # Si no se especifica continente, devuelve la correlación de 'Life Ladder' con todas las variables para todos los continentes
        results = []
        continentes = df['continente'].unique()
        
        for continente in continentes:
            # Filtra el DataFrame por continente
            df_continente = df[df['continente'] == continente]
            
            # Selecciona las columnas numéricas que queremos usar (excluyendo 'Country name' y 'year')
            columns = ['Life Ladder', 'Log GDP per capita', 'Social support', 
                       'Healthy life expectancy at birth', 'Freedom to make life choices', 
                       'Generosity', 'Perceptions of corruption', 
                       'Positive affect', 'Negative affect']
            
            # Asegúrate de que todas las columnas numéricas existan en el DataFrame filtrado
            if all(col in df_continente.columns for col in columns):
                # Calcula las correlaciones de 'Life Ladder' con todas las demás variables
                correlations = df_continente[columns].corr().loc['Life Ladder']
                
                # Convierte a formato largo para facilitar la visualización
                correlations = correlations.reset_index()
                correlations.columns = ['Variable', 'Correlation']
                correlations['continente'] = continente
                
                # Añadir los resultados a la lista
                results.append(correlations)
            else:
                print(f"Algunas columnas no están presentes en el DataFrame del continentee {continente}.")
        
        # Combina todos los resultados en una sola tabla
        combined_results = pd.concat(results, ignore_index=True)
        
        return combined_results
```

```python
def matriz_correlacion_pais(df, country_name=None):
    # Seleccionar las columnas numéricas de interés
    columns = ['Life Ladder', 'Log GDP per capita', 'Social support', 
               'Healthy life expectancy at birth', 'Freedom to make life choices', 
               'Generosity', 'Perceptions of corruption', 
               'Positive affect', 'Negative affect']
    
    if country_name:
        # Filtrar el DataFrame por el país especificado
        df_country = df.loc[df['Country name'] == country_name, columns]
        
        # Verificar si hay datos para el país especificado
        if df_country.empty:
            raise ValueError(f"No hay datos para el país '{country_name}'")
        
        # Calcular las correlaciones con 'Life Ladder'
        correlations = df_country.corr().loc['Life Ladder']
        correlations = correlations.reset_index()
        correlations.columns = ['Variable', 'Correlation']
        correlations['Country'] = country_name
        
        return correlations
    
    else:
        # Calcular las correlaciones para todos los países
        results = []
        for country in df['Country name'].unique():
            df_country = df.loc[df['Country name'] == country, columns]
            
            # Verificar si hay datos para el país
            if not df_country.empty:
                correlations = df_country.corr().loc['Life Ladder']
                correlations = correlations.reset_index()
                correlations.columns = ['Variable', 'Correlation']
                correlations['Country'] = country
                results.append(correlations)
        
        # Combinar todos los resultados en una sola tabla
        combined_results = pd.concat(results, ignore_index=True)
        return combined_results
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error


features = ['Log GDP per capita', 'Social support', 
            'Healthy life expectancy at birth', 'Freedom to make life choices', 
            'Generosity', 'Perceptions of corruption', 
            'Positive affect', 'Negative affect']

# Separar las variables independientes (X) y la dependiente (y)
X = df_felicidad[features]
y = df_felicidad['Life Ladder']

# Estandarizar las variables independientes
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Crear y ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_scaled, y)

# Obtener los coeficientes
coefficients = pd.DataFrame({
    'Feature': features,
    'Coefficient': model.coef_
})

# Ordenar los coeficientes por su valor absoluto para ver qué variable tiene el mayor efecto
coefficients['Absolute Coefficient'] = coefficients['Coefficient'].abs()
coefficients_sorted = coefficients.sort_values(by='Absolute Coefficient', ascending=False)

print("Coeficientes de las variables predictoras:")
print(coefficients_sorted)

# Evaluar el modelo
y_pred = model.predict(X_scaled)
mse = mean_squared_error(y, y_pred)
print(f"Error cuadrático medio del modelo: {mse:.4f}")
```
